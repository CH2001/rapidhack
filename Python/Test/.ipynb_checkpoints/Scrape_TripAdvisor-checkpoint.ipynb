# Actually Getting Started

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By

from bs4 import BeautifulSoup

import pandas as pd
import time
```

```python
# Open TripAdvisor Website
# Don't close the page that pops up

s = Service("C:/Users/elzzh/Documents/SIT/Academic/Trimester_1/INF1002_ProgrammingFundamentals/jupyter_SIT/chromedriver.exe")
driver = webdriver.Chrome(service = s) # <-- You will be using this variable alot, moving forward
```

```python
url = "https://www.tripadvisor.com.sg/Attraction_Review-g294265-d1837767-Reviews-Sands_Skypark_Observation_Deck-Singapore.html"
driver.get(url)
```

```python
see_more_button = driver.find_element(By.XPATH, "/html/body/div[1]/main/div[1]/div[1]/div[2]/div[2]/div/div[1]/section[10]/div/div/div/section/section/div[1]/div/div[5]/div/button")
see_more_button.click()
```

```python
source = driver.page_source
soup = BeautifulSoup(source)
```

```python
print(soup.prettify())
```

```python
reviews1 = soup.find_all("div", "cFSbV Ra Pk PY Px PK")
print(len(reviews1))

reviews2 = soup.find_all("div", "_c")
print(len(reviews2))
```

```python
review_list = []

for review in reviews1:
    try:
        review_dict = {}
        
        review_dict['name']          = review.find('a', class_="BMQDV _F Gv wSSLS SwZTJ FGwzt ukgoS").text
        review_dict['contributions'] = review.find('div', class_="JINyA").text
        review_dict['tag']           = review.find('div', class_="biGQs _P fiohW qWPrE ncFvv fOtGX").text
        review_dict['review']        = review.find('div', class_="biGQs _P pZUbB KxBGd").text
        review_dict['travel_date']   = review.find_all('span', class_="EUrfs")[0].text
        review_dict['travel_pax']    = review.find_all('span', class_="EUrfs")[1].text
        review_dict['review_date']   =review.find('div', class_="biGQs _P pZUbB ncFvv osNWb").text
    
        review_list.append(review_dict)
        
    except:
        continue

for review in reviews2:
    try:
        review_dict = {}
        
        review_dict['name']          = review.find('a', class_="BMQDV _F Gv wSSLS SwZTJ FGwzt ukgoS").text
        review_dict['contributions'] = review.find_all('div', class_="JINyA")[1].text
        review_dict['tag']           = review.find('div', class_="biGQs _P fiohW qWPrE ncFvv fOtGX").text
        review_dict['review']        = review.find('div', class_="biGQs _P pZUbB KxBGd").text
        review_dict['travel_date']   = review.find('div', class_="RpeCd").text.split("•")[0].strip()
        review_dict['travel_pax']   = review.find('div', class_="RpeCd").text.split("•")[1].strip()
        review_dict['review_date']   =review.find('div', class_="biGQs _P pZUbB ncFvv osNWb").text
        
        review_list.append(review_dict)

    except:
        continue
```

```python
display(review_list)
```

```python
data = pd.DataFrame(review_list)
```
