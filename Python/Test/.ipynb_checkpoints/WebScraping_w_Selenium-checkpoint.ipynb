# Actually Getting Started

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By

from bs4 import BeautifulSoup

import pandas as pd
import time
```

```python
# Open Shopee Website

s = Service("C:/Users/elzzh/Documents/SIT/Academic/Trimester_1/INF1002_ProgrammingFundamentals/jupyter_SIT/chromedriver.exe")
driver = webdriver.Chrome(service = s) # <-- You will be using this variable alot, moving forward
```

```python
driver.get("https://shopee.sg/")
```

```python
Popup_Close_Button = driver.execute_script('return document.querySelector("#main shopee-banner-popup-stateful").shadowRoot.querySelector("div.home-popup__close-area div.shopee-popup__close-btn")')
Popup_Close_Button.click()
```

```python
Search_Bar = driver.find_element(By.XPATH, "/html/body/div[1]/div/header/div[2]/div/div[1]/div[1]/div/form/input")
Search_Bar.send_keys("watch")
```

```python
Search_Button = driver.find_element(By.XPATH, "/html/body/div[1]/div/header/div[2]/div/div[1]/div[1]/button")
Search_Button.click()
```

```python
Username_Bar = driver.find_element(By.XPATH, "/html/body/div[1]/div/div[2]/div/div/div/div[2]/form/div/div[2]/div[2]/div[1]/input")
Password_Bar = driver.find_element(By.XPATH, "/html/body/div[1]/div/div[2]/div/div/div/div[2]/form/div/div[2]/div[3]/div[1]/input")
Login_Button = driver.find_element(By.XPATH, "/html/body/div[1]/div/div[2]/div/div/div/div[2]/form/div/div[2]/button")
```

```python
Username_Bar.send_keys("test")
Password_Bar.send_keys("test_password")
Login_Button.click()
```

```python
page_source = driver.page_source
```

```python
soup = BeautifulSoup(page_source)
print(soup.prettify())
```

```python
items = soup.find_all("div", class_="KMyn8J")

object_list = [] # <-- save each object, into list

for item in items:
    object = {}
    
    name = item.find("div", class_="Cve6sh").text
    price = item.find("span", class_="ZEgDH9").text

    object["name"] = name
    object["price"] = float(price)

    print(object)
    print("")

    object_list.append(object)
```

```python
object_list
```

# Use Pandas, to manipulate data in table form

```python
table = pd.DataFrame(object_list)
```

```python
object_list = [] # <-- save each object, into list

for number in range(1,18):
    url = f"https://shopee.sg/search?is_from_login=true&keyword=watch&page={number}"
    driver.get(url)
    time.sleep(5)
    page_source = driver.page_source
    soup = BeautifulSoup(page_source)
    items = soup.find_all("div", class_="KMyn8J")

    for item in items:
        object = {}
        
        name = item.find("div", class_="Cve6sh").text
        price = item.find("span", class_="ZEgDH9").text
    
        object["name"] = name
        object["price"] = float(price)
    
        print(object)
        print("")
    
        object_list.append(object)
```
